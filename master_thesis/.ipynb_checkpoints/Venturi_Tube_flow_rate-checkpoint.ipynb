```python
# 一次大量讀取cell資料的方法
# hint: 如果xlsx的cell內是公式的話，想要讀取出值，必須在
# 前面 wb = load_workbook('data1.xlsx',data_only=True)加上 data_only=True。
parameter_list_number = []
for row in iso_sheet.iter_cols(min_row=1, min_col=11, max_row=17, max_col=11): #每一個row
    for cell in row: #每一個row裡的cell
        parameter_list_number.append(cell.value)
        #print(cell.value, end=" ")    
print(parameter_list_number)
```

# 1
### P1、P2處之壓力
先算出P1(kpa)、P2(kpa)
xlsx內輸入10筆壓差後，經過計算得到多個參數。

```python
# p1_kpa  (P1壓力)
p1_kpa = [] #從xlsx讀取數值
for row in iso_sheet.iter_cols(min_row=2, min_col=5, max_row=11, max_col=5):
    for cell in row:
        p1_kpa.append(cell.value)

p1_kpa_label = []#建立變數名稱
for i in np.arange(1,11): # 1 2 3~~~10
    p1_kpa_label.append('P'+ repr(i)) # repr()將不同類型的變數串成字串。
print(p1_kpa_label)

#將兩個list變成dict，引用參數 dict_p1_kpi['']
dict_p1_kpa = dict(zip(p1_kpa_label, p1_kpa)) #字典還有另一種方法 a ={'x':y}
print(dict_p1_kpa['P1'])

#方便查看變數，列印出表
# 引用參數迭代，p1_kpa[]
pd.DataFrame(p1_kpa, columns=['P1(kpa)'], index=p1_kpa_label)
```

```python
# p2_kpa  (P2壓力)
p2_kpa = []
for row in iso_sheet.iter_cols(min_row=2, min_col=6, max_row=11, max_col=6):
    for cell in row:
        p2_kpa.append(cell.value)

p2_kpa_label = []
for i in np.arange(1,11):
    p2_kpa_label.append('P'+ repr(i))
print(p2_kpa_label)

#將兩個list變成dict，引用參數 dict_p1_kpi['']    
dict_p2_kpa = dict(zip(p2_kpa_label, p2_kpa))
print(dict_p2_kpa['P1'])

#方便查看變數，列印出表
# 引用參數迭代，p2_kpa[]
pd.DataFrame(p2_kpa,columns=['P2(kpa)'] ,index=p2_kpa_label)
```

```python
# 10筆壓差  delta_P=P2-P1 (pa) 這裡要將 kpa轉成 pa
delta_P = []
for i in np.arange(0,10):
    x = (p1_kpa[i]-p2_kpa[i])*1000
    delta_P.append(x)

delta_P_label = []
for i in np.arange(1,11):
    delta_P_label.append('delta_P'+repr(i))
    
pd.DataFrame(delta_P, index=delta_P_label, columns=['delta_P(壓差)'])
```

# 2   
### ideal gas equation : $$\rho = \frac{P}{R K}$$
$$\rho=rho$$
- [在Jupyter Notebook裡面寫Python程式碼和數學公式](https://www.itread01.com/content/1546065568.html)
- [在Jupyter notebook中編輯公式](https://www.cnblogs.com/laoduan/p/7641698.html)

```python
#p1
rho1 = []
for i in np.arange(0,10): #[0~9]共10個
    x = float(p1_kpa[i])/dic['R']/dic['K']
    rho1.append(x)

rho1_label = []
for i in np.arange(1,11):
    rho1_label.append('p'+repr(i))

pd.DataFrame(rho1, index=rho1_label, columns=['p1空氣密度(kg/m^3)'])
```

```python
#p2
rho2 = []
for i in np.arange(0,10):
    x = float(p2_kpa[i]/dic['R']/dic['K'])
    rho2.append(x)

rho2_label = []
for i in np.arange(1,11):
    rho2_label.append('p'+repr(i))

pd.DataFrame(rho2, index=rho2_label, columns=['p2空氣密度(kg/m^3)'])
```

$$\tau = \frac{\rho2}{\rho1}$$

```python
tau = []
for i in np.arange(0,10): #[0~9]
    x = rho2[i]/rho1[i]
    tau.append(x)

tau_label = []
for i in np.arange(1,11):
    tau_label.append('tau'+repr(i))

pd.DataFrame(tau, index=tau_label, columns=['tau'])
```

$$\varepsilon$$

```python
import math
E = []
for i in np.arange(0,10):
    x1 = float((dic['k']*tau[i]**(2/dic['k']))/(dic['k']-1))
    x2 = float((1-dic['b']**4)/(1-(dic['b']**4)*(tau[i]**(2/dic['k']))))
    x3 = float((1-tau[i]**((dic['k']-1)/dic['k']))/(1-tau[i]))
    E_ans = math.sqrt(x1*x2*x3)
    E.append(E_ans)

E_label = []
for i in np.arange(1,11):
    E_label.append('E'+repr(i))

pd.DataFrame(E, index=E_label, columns=['E'])
```
